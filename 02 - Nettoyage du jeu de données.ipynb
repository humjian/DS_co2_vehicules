# Imports et lecture du fichier

```python
# Imports
import pandas as pd
```

```python
# Lecture du fichier
df = pd.read_csv('Data/17-fic_etiq_edition_mars2014.csv', sep=';', decimal=',',encoding='latin-1')

# Suppression des 4 colonnes qui ne servent à rien à la fin
df = df.drop(df.iloc[:, 26:], axis = 1)
```

# Nettoyage du jeu de données

```python
# Suppression des colonnes lib_mod_doss, lib_mod, cnit et tvv
df = df.drop(['lib_mod_doss', 'lib_mod', 'cnit', 'tvv'], axis = 1)

# Suppression de la colonne dscom (quand on aura compris les incohérences)
# df = df.drop(['dscom'], axis = 1)
```

```python
#cod_cbr => supprimer
df = df.drop(['cod_cbr'], axis = 1)

#hybride => supprimer
df = df.drop(['hybride'], axis = 1)

#puiss_admin_98 => garder, 0 NA
#puiss_max =>garder, 0 NA

#typ_boite_nb_rapp => supprimer
df = df.drop(['typ_boite_nb_rapp'], axis = 1)

#conso_urb => garder, supprime NA 0,08%
df = df.dropna(axis = 0, how = 'any',subset=['conso_urb'])
```

```python
# CO2 : enlever les nan
df = df.dropna(axis = 0, how = 'all', subset = ['co2'])

# co_typ_1 : remplacer par la médiane :
df['co_typ_1'].replace( { np.NaN :df['co_typ_1'].median() } , inplace=True)

# hc : on supprime la variable car 80% des données manquantes
df = df.drop ('hc', axis=1)

# hcnox : à supprimer également car elle est issue en partie de 'hc' (80% de nan)
df = df.drop ('hcnox', axis=1)

# nox : on remplace par la médiane
df['nox'].replace( { np.NaN :df['nox'].median() } , inplace=True)

# Conso mixte : les valeurs manquantes étaient les mêmes que sur la variable co2 > en supprimant les lignes concernées de co2, on n'a plus de valeurs manquantes dans conso mixte
```

```python
# Supression des colonnes 'date de mise à jour' et 'champ_v9'
df = df.drop(['date_maj', 'champ_v9'], axis = 1)

# Renommage des valeurs des catégories qui font doublon
df['Carrosserie'] = df['Carrosserie'].replace(to_replace=['COMBISPCACE', 'MONOSPACE COMPACT'], value=['COMBISPACE', 'MINISPACE'])
df['gamme'] = df['gamme'].replace(to_replace=['MOY-INFERIEURE'], value=['MOY-INFER'])

# Remplacement des valeurs manquantes dans la colonne 'ptcl' par la médiane par type de carrosserie
df['ptcl'] = df.groupby('Carrosserie')['ptcl'].transform(lambda x: x.fillna(x.median()))
```

# Ecriture du fichier propre en csv
